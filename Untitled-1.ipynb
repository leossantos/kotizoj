```python
#get brazilian stock data
import pandas as pd
import yfinance as yf
import numpy as np

def get_stock_price(ticker):
    stock = yf.Ticker(f"{ticker}.SA")
    return stock.fast_info['lastPrice']

def get_stock_data(ticker):
    stock = yf.Ticker(f"{ticker}.SA")
    return stock

def input_stock_price(wallet):
    for stock in wallet:
        if stock['Ativo'] == 'Tesouro Selic':
            continue
        print(stock['Ativo'])
        stock_price = get_stock_price(stock['Ativo'])
        stock['Preço'] = round(stock_price, 2)
```

```python
#explique o código abaixo

df = pd.DataFrame(wallet)
df['Posição'] = df['Preço']*df['Quantidade Atual']
df['Balanceamento Atual'] = df['Posição']/df['Posição'].sum()
atual = df['Posição'].sum()
fator_ajuste = 1/df[df['Comprar'] == True]['Balanceamento'].sum()
sum_sem_aporte = df[df['Comprar']== False]['Balanceamento'].sum()
df['Posição Ideal'] = np.where(
    (df['Comprar']==True) & (df['Balanceamento']*(aporte+atual) > df['Posição']),
    df['Balanceamento']*(aporte+atual),
    df['Posição']
)

df['Aporte Ideal'] = np.where(
    df['Posição Ideal'] > df['Posição'], df['Posição Ideal'] - df['Posição'], 0
) 
sum_sem_aporte = df[df['Comprar']== False]['Balanceamento'].sum()

fator_ajuste = 1/df[df['Aporte Ideal'] > 0]['Balanceamento'].sum()


sum_sobra = (aporte+atual)- df['Posição Ideal'].sum()
df['Posição Ideal'] = np.where(
    df['Aporte Ideal'] > 0,
    df['Posição Ideal'] + df['Balanceamento']*fator_ajuste*sum_sobra,
    df['Posição']
)

df['Aporte Ideal'] = df['Posição Ideal'] - df['Posição']
df['Quantidade Aporte'] = np.where(
    df['Ativo'] != 'Tesouro Selic',
    np.trunc(df['Aporte Ideal']/df['Preço']),
    np.trunc(df['Aporte Ideal']/df['Preço']*100)/100
)

df['Aporte Real'] = df['Quantidade Aporte']*df['Preço']
df['Posição após aporte real'] = df['Posição'] + df['Aporte Real']
df['Balanceamento após aporte real'] = df['Posição após aporte real']/df['Posição após aporte real'].sum()
restante_aporte = aporte - df['Aporte Real'].sum()

wallet = df.to_dict('records')
```
